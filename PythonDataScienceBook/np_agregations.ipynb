```python
import pandas as pd
file = pd.read_csv('/home/lukasz/jupyter_notebook/PythonDataScienceBook/notebooks/data/president_heights.csv', sep=',')
file
```

```python
heights = np.array(file['height(cm)'])
```

```python
print(heights)
```

```python
print('Mean: {}'.format(np.mean(heights)))
print('Standard deviation: {}'.format(np.std(heights)))
print('Minimum height: {}'.format(np.amin(heights)))
print('Maximum height: {}'.format(np.amax(heights)))
```

```python
print('25th percentile: {}'.format(np.percentile(heights, 25)))
print('Median: {}'.format(np.median(heights)))
print('75th percentile: {}'.format(np.percentile(heights, 75)))
```
