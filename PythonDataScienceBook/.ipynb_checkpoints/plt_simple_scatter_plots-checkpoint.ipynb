```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, '-ok');
```

```python
help(plt.plot)
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=12, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
plt.scatter(x, y, marker='o'); # another way of creating scatter points
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis') # each point could be different
plt.colorbar();  # show color scale
```
