```python
import numpy as np
import pandas as pd
```

```python
abbrev = pd.read_csv('/home/lukasz/Pobrane/data-USstates-master/state-abbrevs.csv', sep=',')
area = pd.read_csv('/home/lukasz/Pobrane/data-USstates-master/state-areas.csv', sep=',')
popul = pd.read_csv('/home/lukasz/Pobrane/data-USstates-master/state-population.csv', sep=',')
```

```python
popul
```

```python
merged = pd.merge(abbrev, popul, how='outer', left_on='abbreviation', right_on='state/region')
merged = merged.drop('abbreviation', 1) # drop duplicate column
merged
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()]
```

```python
merged[merged['state'].isnull()]
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
```

```python
merged.isnull().any()
```

```python
final = pd.merge(merged, area, on='state', how='left')
```

```python
final.isnull().any()
```

```python
final.dropna(inplace=True)
final
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
```

```python
density.sort_values(ascending=False, inplace=True)
density.head()
```
